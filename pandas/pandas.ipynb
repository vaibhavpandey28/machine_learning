## extract the episode and create seprate column for it


```python
def extract_episodes(txt):
    """Return the text between the first pair of parentheses in `txt`. If none found, return an empty string."""
    if not isinstance(txt, str):
        return ""
    start = txt.find('(')
    if start == -1:
        return ""
    end = txt.find(')', start+1)
    if end == -1:
        return ""
    ep_list = txt[start+1:end].strip()
    return int(ep_list.replace('eps','').strip())
```

```python
df['Episodes'] = df['Title'].apply(extract_episodes)
```

### make a new column for time stamps

```python
def extract_timestamp(txt):
    """Return the text between the first pair of parentheses in `txt`. If none found, return an empty string."""
    if not isinstance(txt, str):
        return ""
    start = txt.find(')')
    if start == -1:
        return ""
    timestamp = txt[start+1:start + 20].strip()
    return timestamp
```

```python
df['Total_Time'] = df['Title'].apply(extract_timestamp)
df.head()
```

```python
from dateutil.relativedelta import relativedelta
from datetime import datetime

def calculate_total_months(period):
    try:
        start_str, end_str = period.split(' - ')
        start_date = datetime.strptime(start_str, '%b %Y')
        end_date = datetime.strptime(end_str, '%b %Y')
        r = relativedelta(end_date, start_date)
        return r.years * 12 + r.months + 1  # +1 to include the starting month
    except:
        return None

df['Months'] = df['Total_Time'].apply(calculate_total_months)
```
